```python
# Implementa el algoritmo de kmeans y justifica la elección del número de 
# clusters. Usa las variables numéricas.
from sklearn.preprocessing import StandardScaler

numeric_cols = ['Rating', 'Reviews',	'Year', 'Price']
X = df.loc[:, numeric_cols]
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
X_norm = pd.DataFrame(X_norm, columns=numeric_cols)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmax   = 16
grupos = range(2, kmax)
wcss   = []
sil_score = []

for k in grupos:
    model = KMeans(n_clusters=k, random_state = 47)
    clusters = model.fit_predict(X_norm)
    wcss.append(model.inertia_)
    sil_score.append(silhouette_score(X_norm, clusters))
  
fig, axs = plt.subplots(1, 2, figsize=(15, 6))
axs[0].plot(grupos, wcss)
axs[0].set_title('Método del codo')
axs[1].plot(grupos, sil_score)
axs[1].set_title('Silhouette Score')
```

Del gráfico anterior se obtiene k = 4, pues representan por mucho mejor los grupos que otros valores de k

```python
model = KMeans(n_clusters=4, random_state=47)
clusters = model.fit_predict(X_norm)
df['Grupo'] = clusters.astype('str')
```

```python
sns.pairplot(data=df, hue='Grupo', palette='tab10')
plt.suptitle('4 grupos', y=1.05)
```

Analiza las características de cada grupo. ¿Qué nombre le pondrías a cada segmento?

*   Grupo 0, serían los libros más baratos, con mayores reviews, con menor rating
*   Grupo 1, serían los libros más viejos más baratos, con menos reviews con rating medio-alto 
*   Grupo 2, serían los libros con mayor precio, con menos reviews con alto rating
*   Grupo 3, serían los lbros más recientes, con menos reviews, con menor precio con rating medio-alto
